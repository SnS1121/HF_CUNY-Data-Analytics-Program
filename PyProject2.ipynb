```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
df= pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv')
```

```python
# isolating columns I will be working with
df[['race', 'gender', 'age', 'time_in_hospital', 'num_medications', 'number_inpatient', 'diag_1', 'diag_2', 'number_diagnoses', 'A1Cresult', 'diabetesMed', 'readmitted']]
```

```python
#created new df
df2=df[['race', 'gender', 'age', 'time_in_hospital', 'num_medications', 'number_inpatient', 'diag_1', 'diag_2', 'number_diagnoses', 'A1Cresult', 'diabetesMed', 'readmitted']]
```

```python
df2.head()
```

```python
df2.groupby(['race', 'age']).count()
```

```python
df2.groupby(['race', 'age']) ['time_in_hospital'].mean()
```

```python
# data shows that hospital stays increases with age; 80-90 yrs had the highest hospital stay among the age group
df2.groupby(['age'])['time_in_hospital'].mean()
```

```python
#plotted avg # of times in hospital by age group
df2.groupby(['age'])['time_in_hospital'].mean().plot(kind="bar")
```

```python
#African Americans have the most # of times in hospital among the group
df2.groupby(['race'])['time_in_hospital'].mean()
```

```python
#plotted statistic above
df2.groupby(['race'])['time_in_hospital'].mean().plot(kind="bar", x='race', y='time_in_hospital')
```

```python
#filtered until it is customized according to what I needed
df2.query("A1Cresult == '>8' & gender == 'Female' & diabetesMed == 'Yes' & readmitted == '<30' & age == '[50-60)' & race != ('?','Other') & diag_1.str.startswith('250').values & num_medications  ") 
```

```python
#mirrored most of what I queried above: (Female 50-60 yrs old pts with A1C results over 8%, is on diabetes meds, readmitted less than 30 days and has a primary dx of diabetes)
df3 = df2[(df2['A1Cresult'] == '>8') & (df2['gender'] == 'Female') & (df2['diabetesMed'] == 'Yes') & (df2 ['readmitted'] == '<30') & (df2['age'] == '[50-60)') & df2['diag_1'].str.startswith('250').values]
```

```python
df3.head()
```

```python
df3.info()
```

```python
#further filtered to only the columns I want to analyze
df4=df3[['race', 'gender', 'age', 'time_in_hospital', 'num_medications', 'number_inpatient', 'diag_1', 'number_diagnoses', 'A1Cresult', 'diabetesMed', 'readmitted']]
```

```python
df4.info()
```

```python
df4.hist()
```

```python
#customized filter excluded Asian, Unknown and Other population 
# Static Filter: Female 50-60 yrs old pts with A1C results over 8%, is on diabetes meds, readmitted less than 30 days and has a primary dx of diabetes
#data showed there is a slight correlation between # of meds and # of dx
df4.groupby(['num_medications','number_diagnoses']).mean()
```
